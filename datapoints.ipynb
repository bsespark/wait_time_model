```python
import pandas as pd
df = pd.read_csv('df_for_model_v3.csv')
```

```python
# Count number of rows per 5_minute_bucket
bucket_counts = df['relative start (minutes)'].value_counts().sort_index()

# Display counts
print(bucket_counts)

# Optional: nice table
import pandas as pd
bucket_counts_df = bucket_counts.reset_index()
bucket_counts_df.columns = ['relative start (minutes)', 'count']
print(bucket_counts_df)

# Optional: visualize
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(bucket_counts_df['relative start (minutes)'], bucket_counts_df['count'], color='skyblue', edgecolor='black')
plt.xlabel('Minute Bucket (Relative to Event Start)')
plt.ylabel('Number of Data Points')
plt.title('Data Points per Minute Bucket')
plt.axvline(0, color='red', linestyle='--', label='Event Start')
plt.legend()
plt.show()
```
